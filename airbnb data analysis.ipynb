```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("airbnb_data.csv")
```

```python
df.head(3)
```

```python
df.columns
```

# Checking Missing Values

```python
print(df.isnull().sum())
```

```python
df.info()
```

# Missing value handling

```python
df['last review'] = pd.to_datetime(df['last review'],errors='coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month' : 0, 'last review' : df['last review'].min()}, inplace = True)
```

```python
df.dropna(subset = ['NAME','host name'], inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df=df.drop(columns=['house_rules','license'],errors='ignore')
```

```python
df.head()
```

```python
# Removing the $ sign and covert to float
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df.head()
```

# Removing the duplicates

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

# Descriptive statistics

```python
df.describe()
```

# Visualization

# 1.Distribution of listing prices


```python
plt.figure(figsize = (10,6))
sns.histplot(df['price'],bins= 50,kde = True, color='blue')
plt.title('Distribution of listing price')
plt.xlabel(' price $')
plt.ylabel('Frequency')
plt.show()
```

```python
The histogram shows a fairly evn distribution of listing prices across 
different price ranges, indicating no particular concentration of listings
in any specific price range . The KDE line helps to visualize this even 
spread more clearly, cofirming that the dataset contains listings with a
wide variety of prices.
```

# 2.Distribution of rooms

```python
df['room type']
```

```python
plt.figure(figsize = (8,5))
sns.countplot(x = 'room type', data=df, color='lightgreen')
plt.title('Distribution of room type')
plt.xlabel(' room type')
plt.ylabel('count')
plt.show()
```

```python
It shows a clear distribution of the different room types available in airbnb dataset.
The majority of listing is for 'Entire home/apt' and 'Private room', with 'Shared room' 
and 'Hotel room' being much less common. This insight can be useful to understand the 
availability and popularity of different types of rooms in airbnb.
```

#  Distribution across different neighbourhood groups

```python
plt.figure(figsize=(12,8))
sns.countplot(y='neighbourhood group', data=df, color='aqua', order=df['neighbourhood group'].value_counts().index)
plt.title('Number of listing by neibourhood groups')
plt.xlabel(' count')
plt.ylabel('Neighbourhood group')
plt.show()
```

```python
The count plot shows a clear distribution of the
number of listings across different neighbourhood
groups. Manhattan and Brooklyn dominate the listings,
suggesting they are prime locations for Airbnb. Queens,
Bronx, and Staten Island have fewer listings, indicating 
less availability or popularity.
```

# Price vs Room Type

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='room type', y='price', hue='room type', data=df, palette='Set1')
plt.title('Price vs Room Type')
plt.xlabel(' Room Type')
plt.ylabel('Price ($)')
plt.legend(title='Room Type')
plt.show()
```

```python
This box plot provides a detailed view oof how
prices vary across different room types in the Airbnb 
dataset. It shows that while 'Shared room' tends to have
lower prices, 'Private Room', 'Entire home/apt' have 
higher and more varied price ranges. This viualization 
helps in understanding the pricing dynamics for different 
types of accommodations on Airbnb.
```

# Number of reviews over time

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()
plt.figure(figsize=(12,8))
reviews_over_time.plot(kind='line', color='red')
plt.title('Number of reviews over time')
plt.xlabel('Date')
plt.ylabel('Number of reviews')
plt.show()
```

```python
This line plot provides a clear visualization of the 
number of reviews over time. It help to identify trends 
and patterns in review activity, such as periods of high
or low activity. This information can be useful for understanding
the dynamics of user engagement and the popularity of Airbnb
listings over time. 
The significant spikes and drops in reviews might be worth 
further investigation to understand the underlying causes,
such as changes in Airbnb policies, market conditions, or external events.
```
